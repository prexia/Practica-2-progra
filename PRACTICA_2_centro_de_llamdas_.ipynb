Hay que crear un segemento de atencion prioritaria a la seccion premier que es nueva, que tenga una atencion prioritaria  al momento de llamar, tambien se quiere proporcionar el tiempo máximo de espera en la publicidad para atraer a más clientes. Tambien, es necesario saber qué tanto debe de aumentar
el equipo del centro para atender adecuadamente a los clientes.

```python
import random #El random lo colocamos ya que el tiempo de arribo de los clientes varia entre 1 y 3 pero es aleatorio entre ese valores con una probabilidad unifmorme
import statistics #Importamos el statistics para calcular las estadisticas en la parte final
#Ahora analizando los datos que necesitamos creamos las variables y sus valores
#t=tiempo
t_simulacion= 60 * 24  # Simulación de 24 horas (1 día)
t_llegada_minimo= 1  # Minimo tiempo entre llegadas (minutos)
t_llegada_maximo= 3  # Maximo tiempo entre llegadas (minutos)
t_atencion_minimo= 1 # Tiempo mínimo de atención (minutos)
t_atencion_maximo= 81 # Tiempo máximo de atención (minutos)
probabilidad_premier= 1/6  # Probabilidad de ser cliente premier

#Definimos una funcion la cual simula un dia completo de llamadas
def llamadas(num_lineas):
  #Ponemos 0 como valor ya que es el numero de lineas ocupadas en el momento sin embargo se podria cambiar a -1
  #si lo multiplicamos por algun elemento en el rango que sea menor o igual al tiempo actual
    t_actual = 0
    clientes_totales = 0
    clientes_premier_atendidos = 0
    clientes_no_premier_atendidos = 0
    tiempos_espera_premier = []  # Lista para almacenar tiempos de espera de clientes premier
    tiempos_espera_totales = []  # Lista para almacenar tiempos de espera de todos los clientes
    tiempos_atencion_premier = []  # Lista para tiempos de atención premier
    lineas_ocupadas = [0] * num_lineas  # Estado de cada línea (0 = libre, >0 = tiempo restante ocupado)

    while t_actual < t_simulacion: #Creamos un ciclo while donde siempre que el tiempo actual pertenzca al dia
        #Generamos el tiempo de llegada entre cada cliente
        t_interarribo = random.uniform(t_llegada_minimo, t_llegada_maximo) #Colocamos la variable t_interarribo que sea = a un numero generado por la funcion random entre 1 y 3
        t_actual += t_interarribo # por lo visto definimos el tiempo actual haciendo las sumas de los intearribos
        if t_actual >= t_simulacion: #Si el tiempo ingresado es mayor al del todo el dia no nos va a arrojar nada
            break

        clientes_totales += 1
        cliente_premier = random.random() <= probabilidad_premier #Asignamos la probabilidad de que sea un cliente premier con el .random y que esta sea menor o igual a 1/6

        #Asignamos las lineas que hay disponibles como:

        lineas_disponibles = [i for i in range(num_lineas) if lineas_ocupadas[i] <= t_actual]
        #Si existe algun valor i en el valor de la funcion,  si las lineas ocupadas son menores o iguales al tiempo obtenido, entonces la linea va a estar disponible
        if lineas_disponibles:
            # Si hay líneas disponibles, atender al cliente
            linea_elegida = lineas_disponibles[0]
            t_atencion = random.uniform(t_atencion_minimo, t_atencion_maximo) #El tiempo de atencion va a estar dado aleatoriamente entre 1 y 81 como definimos
            lineas_ocupadas[linea_elegida] = t_actual + t_atencion # Y esta linea va a estar ocupada en el tiempo actual mas el tiempo que sea atendida

            #Ahora vamos a registrar la atencion de un cliente premier que son los que nos interesan
            if cliente_premier:
                clientes_premier_atendidos += 1
                tiempos_atencion_premier.append(t_atencion) #Le añadimos el tiempo de atencion random
                tiempos_espera_premier.append(0) #Como el cliente fue atendido el tiempo de espera es 0
            else:
                clientes_no_premier_atendidos += 1 #Si no fue premier pero si fue atendido lo colocaremos en esta variable
        else:
            #Colocamos un else a la altura del primer if ya que aqui vamos a colocar si las lineas no estan disponibles
            t_espera = min(lineas_ocupadas) - t_actual #El tiempo de espera va a ser igual al el tiempo de las lineas ocupadas menos el tiempo actual
            t_atencion = random.uniform(t_llegada_minimo, t_llegada_maximo)#El tiempo de atencion va a ser un numero random entre 1 y 3 minutos
            linea_elegida = lineas_ocupadas.index(min(lineas_ocupadas))
            lineas_ocupadas[linea_elegida] += t_atencion# Y esta linea va a estar ocupada en el tiempo actual mas el tiempo que sea atendida

            # Registrar atención y espera
            tiempos_espera_totales.append(t_espera) #Agregamos el tiempo al tiempo de espera
            if cliente_premier:#Si es cliente premier
                clientes_premier_atendidos += 1 #Lo agregamos
                tiempos_espera_premier.append(t_espera) #Agregamos su tiempo de espera
                tiempos_atencion_premier.append(t_atencion)#Agregamos su tiempo de atencion
            else:
                clientes_no_premier_atendidos += 1 #Si son premier los añadimos aqui


    print(f"Simulación con {num_lineas} líneas:") #Imprimimos el numero de lienas disponibles en cada caso
    print(f"Clientes totales: {clientes_totales}") #Imprimimos los clientes totales que llegan en el dia
    print(f"Clientes premier atendidos: {clientes_premier_atendidos}") #Luego los clientes premier que son atendidos
    print(f"Clientes no premier atendidos: {clientes_no_premier_atendidos}")# Luego los clientes premier que no fueron atendidos

    if tiempos_espera_totales:
        print(f"Tiempo medio de espera (todos los clientes): {statistics.mean(tiempos_espera_totales):.2f} minutos") #Imprimimos el tiempo de espera de todos los clientes
    if tiempos_espera_premier:
        print(f"Tiempo medio de espera (clientes premier): {statistics.mean(tiempos_espera_premier):.2f} minutos")#Imprimimos el tiempo medio de espera de los clientes premier
        print(f"Tiempo máximo de espera (clientes premier): {max(tiempos_espera_premier):.2f} minutos")#Imprimimos el tiempo maximo de espera de los clientes premier
    if tiempos_atencion_premier:
        print(f"Tiempo medio de atención (clientes premier): {statistics.mean(tiempos_atencion_premier):.2f} minutos")#Imprimimos el tiempo medio de espera de los clientes premier
        print(f"Tiempo máximo de atención (clientes premier): {max(tiempos_atencion_premier):.2f} minutos")#Imprimimos el tiempo maximo de espera de los clientes premier

for lineas in range(1,20): # Ponemos un rango de simulacion entre 1 y 19 lineas ya que segun el ejercicio el banco tiene 19 lineas
    llamadas(lineas)
```
